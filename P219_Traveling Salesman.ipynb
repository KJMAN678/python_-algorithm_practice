```python
# N 都市の数
# Aij 都市 i から 都市 j に移動するときのコスト
```

```python
N = int(input())
```

```python
A = []
```

```python
for i in range(N):
    a = list(map(int, input().split()))
    A.append(a)
```

```python
ALL = 1 << N
```

```python
ALL
```

```python
# cost[n][i]: 訪れた都市の集合が n で、最後にいる都市が i であるときのコスト最小値
cost = []
```

```python
for n in range(ALL):
    cost.append([10**100]*N)
```

```python
cost
```

```python
# 初期条件、最初にいるときの地点は n には含めない
cost[0][0] = 0
```

```python
# n で表現される場合に要素 i が含まれるかを判定して True/False を返す関数
def has_bit(n, i):
    return (n & (i << i) > 0)
```

```python
for n in range(ALL):
    for i in range(N):
        # i から j に移動する遷移を試す
        for j in range(N):
            # すでに訪問済みか、同じ年は無視する
            if has_bit(n, j) or i == j:
                continue
            cost[n|(1<<j)][j] = min(cost[n|(1<<j)][j], cost[n][i] + A[i][j])
```

```python
# 全都市を訪問して始点に戻ってくる最小コストが答え
print(cost[ALL-1][0])
```
