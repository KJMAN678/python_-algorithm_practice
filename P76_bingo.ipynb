```python
import random
import numpy as np
import pandas as pd
import time
from IPython.display import display
```

```python
# 選ばれた数字がビンゴカードに書かれているか確認

# ビンゴカード用の数字の配列
bingo_card = np.array(random.sample(set(range(1,100)), 9)).reshape(3, 3).tolist()
bingo_card

display(pd.DataFrame(bingo_card))

# ビンゴ判定用の真偽値の配列
bingo_judge = np.array([False]*9).reshape(3, 3).tolist()
bingo_judge

term = int(input("ビンゴ用の数を表示する回数を入力してください"))

for _ in range(0, term):
    print(f"{_}回目: ", end="")
    
    # 選ばれた数字
    num = np.array(random.sample(set(range(1,100)), 1))[0]
    
    # num がビンゴカードに書かれているかを調べる
    for i in range(0, 3):
        for j in range(0, 3):
            
            if bingo_card[i][j] == num:
                # もしヒットすれば bingo_judge[i][j]をTrueに変換
                bingo_judge[i][j] = True


    # ビンゴを達成しているかどうかを判定する

    # ビンゴ達成かどうかのフラグ
    bingo = False

    # 横一列がそろったかどうかの判定
    for i in range(0, 3):
        if bingo_judge[i][0] and bingo_judge[i][1] and bingo_judge[i][2]:
            bingo = True

    # 縦一列がそろったかどうかの判定
    for i in range(0, 3):
        if bingo_judge[0][i] and bingo_judge[1][i] and bingo_judge[2][i]:
            bingo = True
        
    # 左上から右下にかけて斜めにそろったかどうかの判定
    for i in range(0, 3):
        if bingo_judge[0][0] and bingo_judge[1][1] and bingo_judge[2][2]:
            bingo = True

    # 右上から左下にかけて斜めにそろったかどうかの判定
    for i in range(0, 3):
        if bingo_judge[0][2] and bingo_judge[1][1] and bingo_judge[2][0]:
            bingo = True
            
    if bingo:
        print("ビンゴ！")
        display(pd.DataFrame(bingo_judge))
        break
        
    print(f"今回は{num}, 一致した数字は{np.sum(bingo_judge)}個")
    display(pd.DataFrame(bingo_judge))
    time.sleep(1.5)
```
