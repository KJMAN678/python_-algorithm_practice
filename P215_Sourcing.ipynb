```python
# N 部品の種類
# M 部品のセット販売件数
```

```python
N, M = list(map(int, input().split()))
```

```python
# 1 始まりにするためダミーを入れる。S は整数に直す
S = [0]
C = [0]
```

```python
for i in range(M):
    s, c = input().split()
    s_val = 0
    for j in range(N):
        if s[j] == 'Y':
            s_val |= 1<<j
    S.append(s_val)
    C.append(int(c))
```

```python
S
```

```python
C
```

```python
# 集合としてあり得るものの個数、2**Nでも同じ
ALL = 1<<N
```

```python
ALL
```

```python
# cost[i][n]: セット i まで見てそろった部品の集合が n であるときのコスト最小値
cost = []
INF = 10**100
```

```python
for i in range(M+1):
    cost.append([INF] * ALL)
```

```python
cost
```

```python
# 初期条件
cost[0][0] = 0
```

```python
# i が小さいところから順に計算
for i in range(1, M+1):
    for n in range(ALL):
        # セット i を買わない
        cost[i][n] = min(cost[i][n], cost[i-1][n])
        # セット i を買う
        cost[i][n|S[i]] = min(cost[i][n|S[i]], cost[i-1][n] + C[i])
```

```python
cost
```

```python
# 答えは部品が全部そろっている状態の最小コスト
# ただし INF のままなら、部品をそろえることは不可能
ans = cost[M][ALL - 1]
```

```python
if ans == INF:
    ans = -1
```
